```python
## capping outliers
update_num_col= ['num_critic_for_reviews','duration','director_facebook_likes','gross',
                 'num_voted_users','cast_total_facebook_likes','num_user_for_reviews','budget',
                 'title_year','movie_facebook_likes']
def outlier_fun(df,co):
    Q1 = df[co].quantile(0.25)
    Q3 = df[co].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    df[co]=np.where(
        df[co] > upper_limit , upper_limit,
        np.where(df[co] < lower_limit, lower_limit, df[co])
    )
```

```python
for i in update_num_col:
    outlier_fun(df,i)
```

```python
## transforming categorical values into label encoded values

le_color = LabelEncoder()
le_genres = LabelEncoder()
le_language = LabelEncoder()
le_country = LabelEncoder()
le_content_rating = LabelEncoder()

x_train['color'] = le_color.fit_transform(x_train['color'])
x_test['color'] = le_color.transform(x_test['color'])

# x_train['genres'] = le_genres.fit_transform(x_train['genres'])
# x_test['genres'] = le_genres.transform(x_test['genres'])

x_train['language'] = le_language.fit_transform(x_train['color'])
x_test['language'] = le_language.transform(x_test['color'])

x_train['country'] = le_country.fit_transform(x_train['color'])
x_test['country'] = le_country.transform(x_test['color'])

x_train['content_rating'] = le_content_rating.fit_transform(x_train['color'])
x_test['content_rating'] = le_content_rating.transform(x_test['color'])
```

```python
## labeling genres column

from sklearn.preprocessing import MultiLabelBinarizer

# x_train['genres'] = x_train['genres'].str.split('|')
# x_test['genres'] = x_test['genres'].str.split('|')


x_train['genres'] = (
    x_train['genres']
    .fillna('')
    .str.split('|')
)

x_test['genres'] = (
    x_test['genres']
    .fillna('')
    .str.split('|')
)


mlb = MultiLabelBinarizer()

X_train_encoded = mlb.fit_transform(x_train['genres'])
X_test_encoded = mlb.transform(x_test['genres'])  

train_genres = pd.DataFrame(
    X_train_encoded,
    columns=mlb.classes_,
    index=x_train.index
)

test_genres = pd.DataFrame(
    X_test_encoded,
    columns=mlb.classes_,
    index=x_test.index
)

x_train_final = pd.concat(
    [x_train.drop(columns=['genres']), train_genres],
    axis=1
)

x_test_final = pd.concat(
    [x_test.drop(columns=['genres']), test_genres],
    axis=1
)


```
